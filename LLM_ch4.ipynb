```python
from openai import OpenAI
from pydantic import BaseModel

client = OpenAI()

class CalendarEvent(BaseModel):
    name: str
    date: str
    participants: list[str]

response = client.responses.parse(
    model="gpt-4o-mini",
    input=[
        {"role": "system", "content": "Extract the event information."},
        {
            "role": "user",
            "content": "Alice and Bob are going to a science fair on Friday.",
        },
    ],
    text_format=CalendarEvent,
)

event = response.output_parsed

print(event)
```

```python
import os
from openai import OpenAI

client = OpenAI()

def create_fewshot_conversation():

    conversation = client.conversations.create(
        metadata={"topic": "review-classifier-demo"},
        items=[
            # 1) 역할 / 규칙
            {
                "type": "message",
                "role": "developer",
                "content": (
                    "너는 한국어 제품 리뷰를 분석하는 분류기야.\n"
                    "항상 JSON만 반환해야 한다.\n"
                    "- sentiment: positive | negative | neutral 중 하나\n"
                    "- summary: 리뷰 내용을 한 문장으로 한국어 요약\n"
                    "추가 설명 텍스트는 절대 쓰지 마."
                ),
            },

            # # 2) Few-shot 예시 1 (긍정)
            # {
            #     "type": "message",
            #     "role": "user",
            #     "content": "배송도 빠르고 품질도 좋아서 너무 만족스러워요.",
            # },
            # {
            #     "type": "message",
            #     "role": "assistant",
            #     "content": (
            #         '{"sentiment": "positive", '
            #         '"summary": "빠른 배송과 좋은 품질에 매우 만족한다."}'
            #     ),
            # },

            # # 3) Few-shot 예시 2 (부정)
            # {
            #     "type": "message",
            #     "role": "user",
            #     "content": "포장이 엉망이고 제품에 흠집이 있어서 정말 실망했습니다.",
            # },
            # {
            #     "type": "message",
            #     "role": "assistant",
            #     "content": (
            #         '{"sentiment": "negative", '
            #         '"summary": "불량 포장과 흠집 때문에 제품에 실망했다."}'
            #     ),
            # },
        ],
    )

    print("생성된 Conversation ID:", conversation.id)
    return conversation.id


def classify_review(conversation_id: str, review: str):
    response = client.responses.create(
        model="gpt-4o-mini",
        conversation=conversation_id,
        input=review,
    )

    # 편의상 전체 텍스트를 합쳐서 보여주는 필드 (JSON 문자열이 들어올 것)
    print("입력 리뷰:", review)
    print("모델 출력(JSON 문자열):", response.output_text)
    print("-" * 60)


if __name__ == "__main__":
    conv_id = create_fewshot_conversation()

    test_reviews = [
        "가격은 좀 비싸지만 디자인이 깔끔하고 성능이 좋아요.",
        "설명서도 없고 구성품이 빠져 있어서 최악이었어요.",
        "음… 그냥 무난무난합니다. 특별히 좋지도 나쁘지도 않아요.",
    ]

    for r in test_reviews:
        classify_review(conv_id, r)
```
