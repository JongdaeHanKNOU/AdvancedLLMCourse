```python
from openai import OpenAI
import os
from google.colab import userdata
os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")

client = OpenAI()
```

```python

resp = client.responses.create(
    model="gpt-4.1-mini",
    tools=[{
        "type": "mcp",
        "server_label": "dmcp",
        "server_description": "Dungeons & Dragons MCP server for dice rolling",
        "server_url": "https://dmcp-server.deno.dev/sse",
        "require_approval": "never",  # 데모용: 승인 없이 바로 호출
    }],
    input="2d4+1 을 굴리고, 합계가 몇인지 알려줘.",
)

print("=== output_text ===")
print(resp.output_text)
print(resp.output)

print("\n=== raw output types ===")
for item in resp.output:
    print(item.type)
```
