```python
from openai import OpenAI
import math
import os
from google.colab import userdata
os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")

client = OpenAI()

response = client.responses.create(
  model="gpt-4o-mini",
  input="점심 메뉴 하나만 추천해줘.",
  top_logprobs=5,
  temperature = 0.2,
  include= ["message.output_text.logprobs"],
)
print(response.output_text)
probs = response.output[0].content[0].logprobs
print(probs)
for tokenprob in probs:
  print("토큰: {}, 로그확률: {:.3f}, 확률: {:.3f}".format(tokenprob.token, tokenprob.logprob, math.exp(tokenprob.logprob)))
  for candidate in tokenprob.top_logprobs:
    print("\t후보 토큰: {}, 로그확률: {:.3f}, 확률: {:.3f}".format(candidate.token, candidate.logprob, math.exp(candidate.logprob)))


```

```python
from openai import OpenAI

import os
from google.colab import userdata
os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")

client = OpenAI()

stream = client.responses.create(
  model="gpt-4o-mini",
  input="혼자서 개발해볼 수 있는, 재미있는 사이드 프로젝트 아이디어를 5개만 알려줘.",
  stream=True
)

for event in stream:
  # print(event.type)
  if event.type == "response.output_text.delta":
    print(event.delta, flush=True, end="")
  elif event.type == "response.output_text.done":
      print()
  elif event.type == "response.done":
      break
```

헬퍼 함수

```python
import json

client = OpenAI()

def analyze_ticket(
    message: str,
    instructions: str,
    temperature: float = 0.7,
    top_p: float = 1.0,
    max_output_tokens: int = 256,
    top_logprobs: int | None = None,
    include_logprobs: bool = False,
):
    kwargs = dict(
        model="gpt-4o-mini",
        instructions=instructions,
        input=message,
        temperature=temperature,
        top_p=top_p,
        max_output_tokens=max_output_tokens,
    )
    if top_logprobs is not None:
        kwargs["top_logprobs"] = top_logprobs  # 0~20 사이 정수
    if include_logprobs:
        # logprobs 데이터를 응답에 포함
        kwargs["include"] = ["message.output_text.logprobs"]  #

    response = client.responses.create(**kwargs)

    print("=== 원문 응답(output_text) ===")
    print(response.output_text)
    print()

    return response
```

초기 시도

```python
basic_instructions = """
너는 SaaS 서비스 고객센터 상담사야.
사용자의 문의를 읽고, 어떤 내용인지 간단히 분류하고 한 줄로 요약해 줘.
카테고리 이름은 자유롭게 정해도 된다.
"""

sample_ticket = """
어제 결제했는데 카드에서 두 번 결제가 된 것 같아요.
영수증에는 한 번만 찍혀 있는데 카드사 앱에는 두 건이 떠 있어서 확인 부탁드립니다.
"""

resp_basic = analyze_ticket(
    message=sample_ticket,
    instructions=basic_instructions,
    temperature=1.0,
    top_p=1.0,
    max_output_tokens=200,
)
```

프롬프트/파라미터 튜닝


```python
better_instructions = """
너는 SaaS 서비스 고객센터 티켓을 분류하고 요약하는 AI 어시스턴트다.

역할:
- 사용자의 문의를 읽고, 미리 정의된 카테고리 중 하나로 분류하고
- 상담사가 빠르게 이해할 수 있도록 한국어 한 문장으로 요약한다.
- 또한 티켓의 긴급도를 1~5 사이 정수로 추정한다. (5가 가장 긴급)

카테고리 목록:
- "billing": 결제/환불/요금 관련
- "bug": 오류/버그/접속 문제
- "feature_request": 기능 추가/개선 요청
- "account": 로그인/비밀번호/계정 정보 관련
- "other": 위에 해당하지 않는 기타 문의

출력 형식:
- 반드시 아래 JSON 형식만 출력한다. 추가 설명/자연어 문장은 넣지 마라.
- summary는 한국어 한 문장, 40자 이내로 작성한다.

{
  "category": "billing",
  "summary": "여기에 한국어 한 줄 요약",
  "priority": 3
}
"""

resp_better = analyze_ticket(
    message=sample_ticket,
    instructions=better_instructions,
    temperature=0.2,        # 분류/요약은 안정성이 중요 → 낮게
    top_p=1.0,
    max_output_tokens=120,  # 요약이 길어지지 않게 상한 조절
)
```

검증/분석

```python
category_only_instructions = """
너는 고객 문의를 읽고 카테고리를 분류하는 분류기다.

카테고리 목록:
- "billing"
- "bug"
- "feature_request"
- "account"
- "other"

규칙:
- 위 카테고리 문자열 중 하나만 출력해라.
- 따옴표나 마침표 등 다른 문자는 넣지 마라.
"""
resp_logprobs = analyze_ticket(
    message=sample_ticket,
    instructions=category_only_instructions,
    temperature=0.7,             # 가장 확률 높은 토큰만 선택
    top_p=1.0,
    max_output_tokens=20,
    top_logprobs=5,
    include_logprobs=True,
)
probs = resp_logprobs.output[0].content[0].logprobs
print(probs)
for tokenprob in probs:
  print("토큰: {}, 로그확률: {:.3f}, 확률: {:.3f}".format(tokenprob.token, tokenprob.logprob, math.exp(tokenprob.logprob)))
  for candidate in tokenprob.top_logprobs:
    print("\t후보 토큰: {}, 로그확률: {:.3f}, 확률: {:.3f}".format(candidate.token, candidate.logprob, math.exp(candidate.logprob)))


```
